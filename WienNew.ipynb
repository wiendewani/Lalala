# No.1
Merek mobil apa saja yang tersedia dan ada berapa banyak mobil untuk tiap merek

```python
df['merk'] = df['Name'].str.split(' ').str[0]
```

```python
df
```

```python
display(df['merk'].value_counts())
```

```python
df_merk = df['merk'].replace('Isuzu', 'ISUZU')
display(df_merk.value_counts())
```

```python
plt.style.use('ggplot')
ax = df_merk.value_counts().plot(kind='barh', figsize=(10,7),
                                        color="coral", fontsize=13);
ax.set_alpha(0.8)
ax.set_title("banyak mobil untuk tiap merek", fontsize=18)
ax.set_xlabel("Jumlah Mobil", fontsize=18)

for i in ax.patches:
    # get_width pulls left or right; get_y pushes up or down
    ax.text(i.get_width()+.3, i.get_y()+.38, \
            str(round((i.get_width()))))

```

```python
Merek mobil yang tersedia berjumlah 31 merek. Merek mobil dan berapa banyak mobil untuk tiap merek tersebut bisa diliat pada grafik diatas. Jumlah mobil paling banyak adalah merek Maruti yang berjumlah 1211 mobil sedangkan yang yang sedikit digunakan adalah merek Bentley, Lamborghini, Ambassador, dan Smart yang hanya berjumlah 1 mobil.
```

# No 4
Berapa banyak mobil yang memiliki total jarak pemakaian di bawah 100.000 kilometer?


```python
tjarak = (df['Kilometers_Driven'] < 100000).value_counts()
```

```python
tjarak
```

```python

    
plt.style.use('ggplot')
ax = tjarak.plot(kind='bar', figsize=(10,7),color="crimson", fontsize=13);
ax.set_alpha(0.8)
ax.set_title("total jarak pemakaian", fontsize=18)
plt.xticks([0,1],['Di bawah 100.000 km','Di atas 100.000 km'], rotation = "horizontal")

for i in ax.patches:
    _x = i.get_x() + i.get_width() / 2
    _y = i.get_y() + i.get_height()
    value = '{}'.format(int(i.get_height()))
    ax.text(_x, _y, value, ha="center", fontsize =15) 
```

Jumlah mobil yang memiliki total jarak pemakaian dibawah 100.000 km yaitu 5.470 mobil. Jumlah mobil yang memiliki total jarak pemakaian di atas 100.000 km yaitu 549 mobil. Sehingga jumlah mobil yang pemakaian di bawah 100.000 km lebih banyak dibandingkan dengan pemakaian di atas 100.000 km
